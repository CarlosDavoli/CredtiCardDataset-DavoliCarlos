con el fin de realizar un analisis sobre los datos existentes eliminaremos los datos desconocidos. - se evalua si utilizarla o no

```python
# Function to calculate missing values by column# Funct
def missing_values_table(df):

  for cols in df:
    # Reemplazar 'Unknown' por NaN
    df[cols] = df[cols].replace('Unknown', np.nan)
    # # Eliminar filas con NaN en esa columna
    # df_creditCard.dropna(subset=[cols], inplace=True)

  # Total missing values
  mis_val = df.isnull().sum()

  # Percentage of missing values
  mis_val_percent = 100 * df.isnull().sum() / len(df)

  # Make a table with the results
  mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

  # Rename the columns
  mis_val_table_ren_columns = mis_val_table.rename(
  columns = {0 : 'Missing Values', 1 : '% of Total Values'})

  # Sort the table by percentage of missing descending
  mis_val_table_ren_columns = mis_val_table_ren_columns[
      mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
  '% of Total Values', ascending=False).round(1)

  # Print some summary information
  print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"
      "There are " + str(mis_val_table_ren_columns.shape[0]) +
        " columns that have missing values.")

  # Return the dataframe with missing information
  return mis_val_table_ren_columns
```

## Implementado el modelo - Random Forest

```python
y = df_creditCard.pop("Attrition_Flag")
X = df_creditCard[['Customer_Age', 'Dependent_count', 'Months_on_book', 'Total_Relationship_Count',
            'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
            'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1',
            'Avg_Utilization_Ratio',]]
#usaremos un train size de 0.9 ya que al momento de usar kfold tendremos en cuenta 10 segmentos (n=10)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, random_state=69)
```

###Normalizacion de datos

```python

scaler = MinMaxScaler()
# el fit solo se hace para los datos de entrenamiento para evitar el data leakage y el transform es para normalizar los datos donde no hay problemas porque son variables continuas
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

###Selección de hiperparametros

```python


# Definimos los hiperparámetros que queremos ajustar y los valores que queremos probar
param_grid = {
    #n_estimators: Representa el número de árboles de decisión
    'n_estimators': [60, 80, 100],
    #max_depth: Especifica la profundidad máxima que puede alcanzar cada árbol en el bosque
    'max_depth': [None],
    #min_samples_split: Indica el número mínimo de muestras requeridas en un nodo para que se produzca una división adicional
    'min_samples_split': [6, 9, 8],
    #min_samples_leaf: Especifica el número mínimo de muestras requeridas en una hoja
    'min_samples_leaf': [2, 3, 5],
}

# Creamos un clasificador de Bosque de Decisión
rf = RandomForestClassifier()

# Realizamos la búsqueda exhaustiva de cuadrícula
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)
grid_search.fit(X, y)

# Mostramos los mejores hiperparámetros encontrados
print("Mejores hiperparámetros encontrados:")
print(grid_search.best_params_)
```

Otros parametros a consultar max_features: Define la cantidad máxima de características que se considerarán al buscar la mejor división en un nodo. Un valor más bajo reduce la complejidad del modelo y puede ayudar a evitar el sobreajuste.

criterion: Indica el criterio utilizado para medir la calidad de una división. Los criterios comunes son "gini" para el índice de Gini y "entropy" para la ganancia de información. Ambos miden la homogeneidad de las clases en los nodos hijos después de la división.

bootstrap: Es un booleano que indica si se deben usar muestras bootstrap para entrenar árboles individuales. Bootstrap es un método de muestreo con reemplazo que se utiliza para construir cada árbol con un subconjunto aleatorio de datos.

```python
# Entrenar el modelo con los datos de entrenamiento
clf.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = clf.predict(X_test)

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo: {:.2f}%".format(accuracy * 100))
```

### Validacion cruzada con K-Folds


```python
#toman los valores numéricos
X_values = X.values
y_values = y.values
```

```python

# Definimos el número de pliegues para K-Folds
k_folds = 10

# Inicializamos el objeto KFold
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

# Listas para almacenar las puntuaciones de precisión en cada iteración
accuracies = []
precisions = []
recalls = []
f1_scores = []

# Realizamos la validación cruzada
for train_index, test_index in kf.split(df_creditCard):
    X_train, X_test = X_values[train_index], X_values[test_index]
    y_train, y_test = y_values[train_index], y_values[test_index]

    # Entrenamos el modelo
    clf.fit(X_train, y_train)

    # Realizamos predicciones en el conjunto de prueba
    y_pred = clf.predict(X_test)

    # Calculamos la precisión y la almacenamos en la lista
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    #agregando metricas
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)

# Calculamos el promedio de las puntuaciones de precisión
mean_accuracy = np.mean(accuracies)
mean_precisions = np.mean(precisions)
mean_recalls = np.mean(recalls)
mean_f1_scores = np.mean(f1_scores)

# Imprimimos los resultados
print("Puntuaciones de precisión en cada pliegue:", accuracies)
print("Muestras clasificadas correctamente sobre el total de muestras:", mean_accuracy)
print("Puntuaciones de precisión en cada pliegue:", precisions)
print("Muestras positivas clasificadas correctamente:", mean_precisions)
print("Puntuaciones de precisión en cada pliegue:", recalls)
print("tasa de verdaderos positivos:", mean_recalls)
print("Puntuaciones de precisión en cada pliegue:", f1_scores)
print("Desequilibrio entre las clases:", mean_f1_scores)
```

Como conclusión para este modelo podemos decir que es aceptable teniendo en cuenta que posee un f1 de 87.78% aunque lo mas importante es ver la tasa de verdaderos positivos ya que lo que se busca es la exactitud en clientes que puedan llegar a dejar la empresa.

```python
# Definimos el número de pliegues para K-Folds
k_folds = 10

# Inicializamos el objeto KFold
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

# Listas para almacenar las puntuaciones de precisión en cada iteración
accuracies = []
precisions = []
recalls = []
f1_scores = []

# Realizamos la validación cruzada
for train_index, test_index in kf.split(df_creditCard):
    X_train, X_test = X_values[train_index], X_values[test_index]
    y_train, y_test = y_values[train_index], y_values[test_index]

    # Entrenamos el modelo
    xgb_model.fit(X_train, y_train)

    # Realizamos predicciones en el conjunto de prueba
    y_pred = xgb_model.predict(X_test)

    # Calculamos la precisión y la almacenamos en la lista
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    #agregando metricas
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)

# Calculamos el promedio de las puntuaciones de precisión
mean_accuracy = np.mean(accuracies)
mean_precisions = np.mean(precisions)
mean_recalls = np.mean(recalls)
mean_f1_scores = np.mean(f1_scores)

# Imprimimos los resultados
print("Puntuaciones de precisión en cada pliegue:", accuracies)
print("muestras clasificadas correctamente sobre el total de muestras:", mean_accuracy)
print("Puntuaciones de precisión en cada pliegue:", precisions)
print("muestras positivas clasificadas correctamente:", mean_precisions)
print("Puntuaciones de precisión en cada pliegue:", recalls)
print("tasa de verdaderos positivos:", mean_recalls)
print("Puntuaciones de precisión en cada pliegue:", f1_scores)
print("desequilibrio entre las clases:", mean_f1_scores)
```
